```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_local = pd.read_csv('./data/city_data.csv')
df_global = pd.read_csv('./data/global_data.csv')
```

```python
df_local.head()
```

```python
df_global.head()
```

```python
df_local.isnull().sum()
```

```python
df_local.dropna(inplace=True)
```

```python
df_global.isnull().sum()
```

```python
df_local.duplicated().sum()
```

```python
df_global.duplicated().sum()
```

```python
df_local.drop(['country'], axis=1, inplace=True)
```

```python
df_local.head()
```

```python
local_temp = np.repeat('local', df_local.shape[0])
global_temp = np.repeat('global', df_global.shape[0])
```

```python
df_local['local'] = local_temp
df_local.head()
```

```python
df_global['global'] = global_temp
df_global.head()
```

```python
temp_df = df_local.append(df_global)
temp_df.head()
```

```python
X = temp_df['year']
Y = temp_df['avg_temp']
plt.rcParams["figure.figsize"] = (20, 10)
plt.title('local city temperature')
plt.xlabel('Year')
plt.ylabel('avg_temp')
plt.plot(X,Y, linestyle='--')
plt.show()
```
